##3
3สร้างตารางจากผลการลงคะแนนเสียงของพรรคที่แสดงอัตราส่วนระหว่างจำนวนเสียงที่ได้รับโดยพรรคที่ได้รับมากที่สุดและพรรคที่ได้รับมากที่สองของแต่ละพื้นที่เขตเลือกตั้ง



```python
merged3_df = RCC_df.merge(RCS_df,left_on='cons_id',right_on='cons_id')
```

```python
merged3_df = merged3_df.merge(IP_df,left_on='prov_id',right_on='prov_id')
```

```python
merged3_df = merged3_df.merge(IPO_df,left_on='party_id',right_on='party_id')
```

```python
merged3_df.head()
```

```python
merged3_df = merged3_df[['cons_id','mp_app_id','province','mp_app_vote','name']]
merged3_df.head()
```

```python
top_2 = merged3_df.groupby('cons_id').apply(lambda x: x.nlargest(2, 'mp_app_vote')).reset_index(drop=True)

top_2 = top_2.sort_values(['cons_id', 'mp_app_vote'], ascending=[True, False])

top_2['V1'] = top_2.groupby('cons_id')['mp_app_vote'].transform('first')
top_2['V2'] = top_2.groupby('cons_id')['mp_app_vote'].transform(lambda x: x.nlargest(2).iloc[-1])

top_2['V1/V2'] = top_2['V1'] / top_2['V2']

top_2 = top_2.drop(columns=['V1', 'V2'])

Ans_3 = top_2
```
